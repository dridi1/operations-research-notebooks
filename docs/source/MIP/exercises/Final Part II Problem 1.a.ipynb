# Final Exam (Solved)
## Part II
### Problem 1
You need to design the distribution network of a company that has several production plans and manufactures different types of products.
You have build a mathematical programming model to optimise your distribution network design.
You have to use the following data in your model:

- the transportation cost per unit of product and product type from a production plan to a warehouse
- the transportation cost per unit of product and product type from a warehouse to a retail region
- the handling costs per unit of product at every warehouse
- the yearly fixed operation costs of every warehouse
- The yearly demand for every product at every retail region
- The maximum production capacity of every production plant for every product

Your company policy states that you must satisfy the demand of every region from a single warehouse.
Your company also states that you cannot use more than C different warehouses. 

Write down:

**a.** The indices, input data and decision variables of your model (1 point)
The indices to use are: 

- i: production plans
- j: warehouses
- k: retail regions
- l: product types

The input data are:
- $S_{ijl}$: Transportation cost per unit of product l from production plan i to warehouse j
- $T_{jkl}$: Transportation cost per unit of product l from warehouse j to retail region k
- $H_{jl}$: Handling cost per unit of product l in warehouse j 
- $F_j$: Yearly fixed operation costs of warehouse j
- $D_{kl}$: Yearly demand of product l at retail region k
- $M_{il}$: Maximum production capacity of production plant i for product l

The decision variables are:

- $X_{ijl}$: Units of product l to transport from production plan i to warehouse j 
- $Y_{jkl}$: Units of product l to transport from warehouse j to retail region k
- $W_{jk}$: binary {1 if the warehouse j supplies the retailer region k } {0 otherwise}
- $I_{j}$: Binary decision variable that represents the selection of a warehouse j {1: we use a warehouse, 0: otherwise}

**b.** The objective function (1 point)

- Transportation costs from production plans to warehouses

$C_{t1} = \sum_i{\sum_j{\sum_l{S_{ijl}·X_{ijl}}}}$

- Transportation cost from warehouses to retail regions

$C_{t2} = \sum_j{\sum_k{\sum_l{T_{jkl}·Y_{jkl}}}}$

- Handling costs

$C_h = \sum_j{\sum_k{\sum_l{H_{jl}·Y_{jkl}}}} $

- Yearly fixed operation costs

$C_f = \sum_j{I_{j}·F_{j}} $


- Objective function
$\min z = C_{t1} + C_{t2} + C_h + C_f$

$
\min z = \sum_i{\sum_j{\sum_l{S_{ijl}·X_{ijl}}}} + \sum_j{\sum_k{\sum_l{T_{jkl}·Y_{jkl}}}} + \sum_j{\sum_k{\sum_l{H_{jl}·Y_{jkl}}}} + \sum_j{I_{j}·F_{j}}$

**c.** The constraints (2 points)

- Demands:
The demand in every region and for every product must be met:

$\sum_{j}{Y_{jkl}}= D_{kl} \forall k, \forall l$

- Material flows:
The amount of each product that leaves a warehouse cannot be higher than the amount of product that enters a warehouse. Assuming that we do not storing any amount for next periods, then both variables are equal instead of less or equal:

$\sum_{i}{X_{ijl}} = \sum_{k}{Y_{jkl}} \forall j, \forall l$

- Production capacities:
The amount of a each product produced in a plant must not exceed production capacity:

$\sum_{j}{X_{ijl}} \leq M_{il} \forall i, \forall l$


- Company Policy constraint:
The number of warehouses must not exceed the limit:

$\sum_{j}{I_{j}} \leq C$

- Single Source Constraint: 
Each region receives all its products from only one warehouse:

$\sum_{j}{W_{jk}} = 1 \forall k$

- Logical constraints: 
Finally, it is necessary to introduce some logical constraints. First, a warehouse must be built if it supplies to any region.

$ W_{jk} \leq I_{j} \forall j, \forall k$

Since both decision variables are binary, with this constraint we make sure that if any warehouse supplies to a region, i.e. if any $W_{jk}$ for any value of k is 1, then we must build a warehouse in that location, i.e. $Y_{j}$ must be 1.


Similarly, a warehouse will supply a region only when the amount transported of all products from such warehouse to such region is non zero.

$\sum_{l}Y_{jkl} \leq M·W_{jk} \forall j, \forall k$

where M is a very large number
